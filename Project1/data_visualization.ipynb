# 数据可视化

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 假设data已经被正确加载
data = pd.read_csv('filled_data.csv')

# 选择特征和标签
features = ['SO2', 'NO', 'NO2', 'NOX', 'O3', 'temperature', 'wind_speed', 'relative_humidity', 'air_pressure']
label = 'PM2_5'

# 创建一个包含所有子图的画幅
fig, axs = plt.subplots(len(features) + 1, 1, figsize=(10, 4*(len(features) + 1)))

# 绘制每个特征的图表
for i, feature in enumerate(features):
    axs[i].plot(data[feature], label=feature)
    axs[i].set_title(feature)
    axs[i].legend()

# 绘制标签的图表
axs[len(features)].plot(data[label], label='PM2.5', color='red')
axs[len(features)].set_title('PM2.5')
axs[len(features)].legend()

plt.tight_layout()
plt.show()
```
